```python
#importing the libraries
import pandas as pd
import numpy as np
```

```python
#importing the dataset 
df = pd.concat(pd.read_excel('Comments.xlsx', sheet_name=None), ignore_index=True)
df
```

```python
df.head(20)
```

```python
df=df.drop(['Name','Time','Reply Count'],axis =1)
df.head(20)
```

```python
type(df['Comment'])
```

```python
df.drop_duplicates('Comment',inplace=True)
df['Comment'][:20]
```

```python
df['Comment'] = df['Comment'].str.lower()
df['Comment'][:20]
```

```python
!pip install demoji
import demoji
demoji.download_codes()
def remove_em(text):
  dem=demoji.findall(str(text))
  for item in dem.keys():
    text=text.replace(item,'')
  return text
df['Comment']=df['Comment'].apply(remove_em)
df['Comment'][:20]
```

```python
import re
def remove_URL(sample):
    """Remove URLs from a sample string"""
    return re.sub(r"http\S+", "", str(sample))
df['Comment']= df['Comment'].apply(remove_URL)
df['Comment'][:20]
```

```python
import re
import string
df['Comment'] = df['Comment'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '' ,str(x)))
df['Comment'][:20]
```

```python
def tokenizer(text):
    return text.split()
df['Comment']=df['Comment'].apply(tokenizer)
df['Comment'][:20]
```

```python
from nltk.stem.snowball import SnowballStemmer
porter=SnowballStemmer("english",ignore_stopwords=False)
def stem_it(text):
    return [porter.stem(word) for word in text]
df['Comment']=df['Comment'].apply(stem_it)
df['Comment'][:20]
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmit_it(text):
    return [lemmatizer.lemmatize(word, pos="a") for word in text]
df['Comment']=df['Comment'].apply(lemmit_it)
df['Comment'][:20]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords 
stop_words = stopwords.words('english')
def stop_it(text):
    review= [word for word in text if not word in stop_words]
    return review
df['Comment']=df['Comment'].apply(stop_it)
df['Comment'][:20]
```
